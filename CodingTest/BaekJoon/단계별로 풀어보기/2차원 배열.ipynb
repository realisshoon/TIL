# 2566 최댓값


### 내가 제출한 답안

- 시간복잡도 O(1)

```python
array99=[list(map(int,input().split())) for _ in range(9)]
max_num=0
row=0
col=0

for i in range(9):
    for j in range(9):
        if max_num<array99[i][j]:
            max_num=array99[i][j]
            col=j
            row=i
        else:
            continue
        
print(max_num)
print(row+1,col+1)
```

### GPT 답안

- 2차원 배열을 1차원으로 flat 작업을 진행함
- 그 다음 최대값을 구하고 그 값에 해당하는 index 를 구함
- index를 9로 나뉘어 몫은 행, 나머지는 열

사용된 `divmod()` `index()` `max()` 외에는 참고 안 해도 될듯 

```python
array99 = [list(map(int, input().split())) for _ in range(9)]

# 모든 값을 1D 리스트로 변환하여 최대값과 인덱스를 찾음
flat_array = [num for row in array99 for num in row]
max_num = max(flat_array)
index = flat_array.index(max_num)

# 위치 계산
row, col = divmod(index, 9)

print(max_num)
print(row + 1, col + 1)
```

# 10798 세로읽기

내가 제출한 답안 : 구현에서 실수함

```python
array=[input() for _ in range(5)]
result=""
max_length=max(len(s) for s in array)

for col in range(max_length):
    for row in range(5):
        if col<len(array[row]):
            result+=array[row][col]

print(result)
```

# 2563 색종이

1. 2차원 배열에 그리는 방법
2. 정석적인 수학 계산 방법

```python
# 1
n=int(input())
grid=[[0]*100 for _ in range(100)]

for _ in range(n):
    x,y=list(map(int,input().split()))
    for i in range(x,x+10):
        for j in range(y,y+10):
            grid[i][j]=1

covered_ared= sum(sum(row) for row in grid)
print(covered_ared)
```
